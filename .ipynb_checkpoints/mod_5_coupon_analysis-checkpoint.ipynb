### Will a Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaraunt near where you are driving. Would you accept that coupon and take a short detour to the restaraunt? Would you accept the coupon but use it on a sunbsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaraunt? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \\$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\\$20 - \\$50). 

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece. 





### Data Description
Keep in mind that these values mentioned below are average values.

The attributes of this data set include:
1. User attributes
    -  Gender: male, female
    -  Age: below 21, 21 to 25, 26 to 30, etc.
    -  Marital Status: single, married partner, unmarried partner, or widowed
    -  Number of children: 0, 1, or more than 1
    -  Education: high school, bachelors degree, associates degree, or graduate degree
    -  Occupation: architecture & engineering, business & financial, etc.
    -  Annual income: less than \\$12500, \\$12500 - \\$24999, \\$25000 - \\$37499, etc.
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she buys takeaway food: 0, less than 1, 1 to 3, 4 to 8 or greater
    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination
    - Location of user, coupon and destination: we provide a map to show the geographical
    location of the user, destination, and the venue, and we mark the distance between each
    two places with time of driving. The user can see whether the venue is in the same
    direction as the destination.
    - Weather: sunny, rainy, or snowy
    - Temperature: 30F, 55F, or 80F
    - Time: 10AM, 2PM, or 6PM
    - Passenger: alone, partner, kid(s), or friend(s)


3. Coupon attributes
    - time before it expires: 2 hours or one day

2. What proportion of bar coupons were accepted?


```python
b0 = df0.query('Y == 0').shape[0]
b1 = df0.query('Y == 1').shape[0]
b2 = df0.shape[0]
x0 = b0/b2
x1 = b1/b2
x2 = round(100*x1,1)


print(f'b0 {b0:8}      num coupons not accepted')
print(f'b1 {b1:8}      num coupons accepted')
print(f'b2 {b2:8}      num total bar coupons')
print(f'x0 {x0:8.3f}      ratio not accepted')
print(f'x1 {x1:8.3f}      ratio accepted')
print(f'x2 {x2:8}% <<< proportion of bar coupons accepted.')
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
df0['bar_le_3x'] = df0['Bar'].apply(lambda x: 1 if x in ('never','less1','1~3') else 0)

srs0 = df0.groupby(['bar_le_3x','Y']).size()
srs1 = df0.groupby(['bar_le_3x']).size()

lst0 = list(srs0)
lst1 = list(srs1)

c0 = lst0[0]
c1 = lst0[1]
c2 = lst0[2]
c3 = lst0[3]
d0 = lst1[0]
d1 = lst1[1]

y0 = c0/d0
y1 = c1/d0
y2 = c2/d1
y3 = c3/d1
print('Note about nomenclature: ')
print('le3 =  drivers who go to bars three or fewer times per month')
print('mt3 =  drivers who go to bars more than three times per month')
print(f'srs0\n{srs0}')
print(f'lst0 {lst0}')
print(f'c0 {c0:8}    num coupons not accepted by mt3 drivers')
print(f'c1 {c1:8}    num coupons     accepted by mt3 drivers')
print(f'c2 {c2:8}    num coupons not accepted by le3 drivers')
print(f'c3 {c3:8}    num coupons     accepted by le3 drivers')
print(f'd0 {d0:8}    num le3 drivers ')
print(f'd1 {d1:8}    num mt3 drivers ')
print(f'y0 {y0:8.3f}     ratio coupons not accepted by mt3 drivers')
print(f'y1 {y1:8.3f}     ratio coupons     accepted by mt3 drivers')
print(f'y2 {y2:8.3f}     ratio coupons not accepted by le3 drivers')
print(f'y3 {y3:8.3f}     ratio coupons     accepted by le3 drivers')
print(f'srs1\n{srs1}')
print(f'lst1\n{lst1}')

```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
df0['bar_mt_1_per_mo'] = df0['Bar'].apply(lambda x: 1 if x in ('1~3','4~8','gt8') else 0)
df0['age_over_25'] = df0['age'].apply(lambda x: 1 if x in ('26','31','36','41','46','50plus') else 0)

age0 = df0.groupby(['bar_mt_1_per_mo','age_over_25','Y']).size()
age1 = df0.groupby(['bar_mt_1_per_mo','age_over_25']).size()

lst2 =list(age0) 
lst3 =list(age1) 
print(f'age0\n{age0}')
print(f'age1\n{age1}')
print(f'lst2\n{lst2}')
print(f'lst3\n{lst3}')
e0 = lst2[0]
e1 = lst2[1]
e2 = lst2[2]
e3 = lst2[3]
e4 = lst2[4]
e5 = lst2[5]
e6 = lst2[6]
e7 = lst2[7]
f0 = lst3[0]
f1 = lst3[1]
f2 = lst3[2]
f3 = lst3[3]
z0 = e0/f0
z1 = e1/f0
z2 = e2/f1
z3 = e3/f1
z4 = e4/f2
z5 = e5/f2
z6 = e6/f3
z7 = e7/f3
z8 = (e0+e2+e4)/(f0+f1+f2)
z9 = (e1+e3+e5)/(f0+f1+f2)
print(f'z0 {z0}')
print(f'z1 {z1}')
print(f'z2 {z2}')
print(f'z3 {z3}')
print(f'z4 {z4}')
print(f'z5 {z5}')
print(f'z6 {z6}   not accept - drivers who go to a bar more than once a month and are over the age of 25')
print(f'z7 {z7}   accept - drivers who go to a bar more than once a month and are over the age of 25 ')
print(f'z8 {z8}   not accept - everyone else')
print(f'z9 {z9}   accept - everyone else')
```

    5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry. 


```python
  
occ0 = df0.groupby(['occupation']).size()
pas0 = df0.groupby(['passanger']).size()


df0['occupation_non_fff'] = df0['occupation'].apply(lambda x: 0 if x in ('Farming Fishing & Forestry') else 1)
df0['pas_non_kid'] = df0['passanger'].apply(lambda x: 0 if x in ('Kid(s)') else 1)

occ1 = df0.groupby(['occupation','occupation_non_fff']).size()
pas1 = df0.groupby(['passanger','pas_non_kid']).size()
occ2 = df0.groupby('occupation_non_fff').size()

#pas2 = df0.groupby('pas_non_kid').sum()#[['pas_non_kid']]
pas2 = df0.groupby('pas_non_kid').size()#[['pas_non_kid']]
pas3 = df0.groupby(['bar_mt_1_per_mo','pas_non_kid','occupation_non_fff','Y']).size()
pas4 = df0.groupby(['bar_mt_1_per_mo','pas_non_kid','occupation_non_fff']).size()

lst4 =list(pas3) 
lst5 =list(pas4) 

#print(f'occ1\n{occ1}')
#print(f'occ2\n{occ2}')
print(f'pas1\n{pas1}')
print(f'pas2\n{pas2}')
print(f'pas3\n{pas3}')
print(f'lst4\n{lst4}')
print(f'lst5\n{lst5}')
g0 = lst4[0]
g1 = lst4[1]
g2 = lst4[2]
g3 = lst4[3]
g4 = lst4[4]
g5 = lst4[5]
g6 = lst4[6]
g7 = lst4[7]
g8 = lst4[8]
g9 = lst4[9]
g10 = lst4[10]
g11 = lst4[11]

h0 = lst5[0]
h1 = lst5[1]
h2 = lst5[2]
h3 = lst5[3]
h4 = lst5[4]
h5 = lst5[5]

w10 = g10/h5
w11 = g11/h5
w12 = (g0+g2+g4+g6+g8)/(h0+h1+h2+h3+h4)
w13 = (g1+g3+g5+g7+g9)/(h0+h1+h2+h3+h4)
print(f'w10 {w10} <<< match criteria no accept coupon')
print(f'w11 {w11} <<< match criteria accept coupon')
print(f'w12 {w12} <<< no criteria no accept coupon')
print(f'w13 {w13} <<< no criteria accept coupon')
print('conclusion') 
print('when criteria match -> 30/70 pct in favor of accepting coupon')
print('when criteria not met opposite is true, namely 70/30 pct in favor of not accepting coupon')
```
